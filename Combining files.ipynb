```python
import pandas as pd
import os

# Specify the file paths of the CSV files
file_paths = [
    '/content/drive/MyDrive/Colab Notebooks/Detecting-RPL-Attacks/Results/DFA.csv',
    '/content/drive/MyDrive/Colab Notebooks/Detecting-RPL-Attacks/Results/SFA.csv',
    '/content/drive/MyDrive/Colab Notebooks/Detecting-RPL-Attacks/Results/SHA.csv',
    '/content/drive/MyDrive/Colab Notebooks/Detecting-RPL-Attacks/Results/SYA.csv',
    '/content/drive/MyDrive/Colab Notebooks/Detecting-RPL-Attacks/Results/VNA.csv'
]

# Create an empty DataFrame to store the merged data
merged_data = pd.DataFrame()

# Iterate over the file paths and read each CSV file
for file_path in file_paths:
    if os.path.exists(file_path):  # Check if the file exists
        # Read the CSV file
        data = pd.read_csv(file_path)
        # Append the data to the merged_data DataFrame
        merged_data = merged_data.append(data, ignore_index=True)
    else:
        print(f"File not found: {file_path}")

# Save the merged data as a new CSV file
output_file = '/content/drive/MyDrive/Colab Notebooks/Detecting-RPL-Attacks/Results/combined-files.csv'
merged_data.to_csv(output_file, index=False)

print("Merging and saving completed.")
```
